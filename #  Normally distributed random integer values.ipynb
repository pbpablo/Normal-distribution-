#  Normally distributed random integer values


```python
import numpy as np
normal_list = np.random.normal(12.5,3,1000)  #the 3 parameters are mean(12.5), standard deviation(3) and size(1000). 
normal_list
```

```python
max(normal_list) #to find the maximum in the list 
```

```python
min(normal_list) #to find the minimun in the list 
```

```python
np.mean(normal_list) #to find mean of the list 
```

```python
import seaborn as sns
x= sns.histplot(data=normal_list, kde=True)
sns.set() #Applies the default seaborn theme, scaling, and color palette.
x.set_xlabel("Normal Number of Random numbers between 0-25", fontsize= 15, color='orange')
x.set_ylabel("Frequency", fontsize= 15, color='orange')
x.set_title("Normal Distribution", fontsize= 15, color='orange')
```
